```python
import pandas as pd
import pickle
from datetime import datetime
```

```python
with open("..\data-dir\data-df.pkl","rb") as fr:
    data = pickle.load(fr)
data
```

```python
RFRD = 0.000084 # 2022-10-14 ~ 2023-01-06 일별 한국무위험지표금리(KOFR)
RFRW = 0.000591 # 2022-10-14 ~ 2023-01-06 일별 한국무위험지표금리(KOFR)
data["Label"] = 0
data
```

```python
data["DayIndex"]
item = data.DayIndex.str.split('_')
data["Datetime"] = item.str.get(0)
data["AMPM"] = item.str.get(1)
data["Datetime"] = pd.to_datetime(data["Datetime"])
data = data.set_index(keys="Datetime")
data.index = pd.to_datetime(data.index)
```

```python
data.index
```

```python
data_weekly_mean = data.resample(rule = "W").mean()
data_weekly_mean
```

```python
mean = list(data_weekly_mean[['Open', 'Close', 'High', 'Low']].mean(axis=1))
mean
```

```python
data = data.reset_index()
data
```

```python
try:
    for i in data.index:
        d = i//7
        if data.loc[i, "Close"] * (1+RFRD) <= data.loc[i+1, "Low"]: # 당일 종가와 내일 저가
            data.loc[i, "Label"] = 2
        elif data.loc[i, "Close"] * (1+RFRW) < mean[d+1]:
            data.loc[i, "Label"] = 1
except (KeyError, IndexError):
    pass
data
```
